* **How many Mozambicans are unemployed?**

```python
Moz= Covid_Impact[(Covid_Impact["EmploymentType"]=="Unemployed") & (Covid_Impact["Country"]=="Mozambique")]
Moz
```

* **How many South Africans received loans?**

```python
Loan= Covid_Impact[(Covid_Impact["COVIDLoans"]=="Yes") & (Covid_Impact["Country"]=="South Africa")]
Loan
```

* **Total number of Nigerians whose top priority was Food.**

```python
Food= Covid_Impact[(Covid_Impact["TopPriority"]=="Food") & (Covid_Impact["Country"]=="Nigeria")]
Food
```

* **How many Kenyans asked their goverment to prioritise reopenening the economy**

```python
Priority= Covid_Impact[(Covid_Impact["Government Priority (What Should Government Prioritise on now)"]=="Reopening the economy") & (Covid_Impact["Country"]=="Kenya")]
Priority
```
